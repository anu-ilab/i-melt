# Model

```julia
# The model

#
# Cps
#

at_gfu(x) = 3.0.*x[1,:] .+ 5.0.*x[2,:] + 3.0.*x[3,:] + 3.0.*x[4,:]# + 2*MgO + 2*MgO
aCpl(x) = 81.37.*x[1,:] .+ 27.21.*x[2,:] .+ 100.6.*x[3,:]  .+ 50.13.*x[4,:] .+ x[1,:].*(x[4,:].*x[4,:]).*151.7

ap(x) = aCpl(x) - 3.0.*8.314.*at_gfu(x)
b(x) = 0.0943.*x[2,:] + 0.01578.*x[4,:] #bCpl

#
# With two subnetworks for Be and Sc
#
m1 = Chain(
  Dense(4, 100, relu),
  Dense(100, 1)) |> gpu

m2 = Chain(
  Dense(4, 50, relu),
  Dense(50, 1)) |> gpu

Be(x) = exp.(m1(x[1:4,:]))
ScTg(x) = exp.(m2(x[1:4,:]))

# Other stuffs

Ae = param([-2.11])

tg(x) = Be(x)./((12.0.-Ae).*ScTg(x))

dCp(x,T, ap, b) = ap.*(log.(T).-log.(tg(x))) .+ b.*(T.-tg(x))

model(x,T, ap, b) = Ae .+ Be(x) ./ (T.* (ScTg(x) .+ dCp(x,T, ap, b)))

mse(yp, y) = sqrt(sum((yp .- y).^2)./size(y, 2))

predict(x,T, ap, b) = model(x,T, ap, b)

# loss functions
loss(x, T, ap, b, y) = mse(model(x, T, ap, b), y)
loss_be(x,be) = mse(Be(x),be)
loss_sc(x,sc) = mse(ScTg(x),sc)
loss_tg(x,target) = mse(tg(x),target)
```

# Training

```julia
function prepare_datas(X_,y_)
    x = X_'[1:4,:]
    x = x./sum(x,dims=1)
    
    y = reshape(y_[:],1,length(y_))

    T = reshape(X_[:,7],1,size(X_,1))
    
    be = reshape(X_[:,8],1,size(X_,1))
    sc = reshape(X_[:,9],1,size(X_,1))
    tg = reshape(X_[:,10],1,size(X_,1))
    return x, y, T, sc, be, tg
end
```

```julia
x_train, y_train, T_train, sc_train, be_train, tg_train = prepare_datas(X_train,y_train)

ap_train = reshape(ap(x_train),1,size(x_train,2))
b_train = reshape(b(x_train),1,size(x_train,2))

x_train = X_train_sc'[1:4,:]
```

```julia
x_valid, y_valid, T_valid, sc_valid, be_valid, tg_valid = prepare_datas(X_valid,y_valid)

ap_valid = reshape(ap(x_valid),1,size(x_valid,2))
b_valid = reshape(b(x_valid),1,size(x_valid,2))

x_valid = X_valid_sc'[1:4,:]
```

```julia
x_test, y_test, T_test, sc_test, be_test, tg_test = prepare_datas(X_test,y_test)

ap_test = reshape(ap(x_test),1,size(x_test,2))
b_test = reshape(b(x_test),1,size(x_test,2))

x_test = X_test_sc'[1:4,:]
```

```julia
record_loss_be_train = Float64[]
record_loss_be_valid = Float64[]

dataset = [(x_train, be_train)]
evalcb = () -> (push!(record_loss_be_train,loss_be(x_train, be_train).data),push!(record_loss_be_valid,loss_be(x_valid, be_valid).data)) #@show(loss_be(x_train, be_train),loss_be(x_valid, be_valid))
opt = ADAM(0.0001)
@epochs 5000 Flux.train!(loss_be, params(m1), dataset, opt, cb = throttle(evalcb, 10))
```

```julia
plot(record_loss_be_valid,label="valid")
plot(record_loss_be_train,label="train")
yscale("log")
legend()
```

```julia
dataset = [(x_train, sc_train)]
evalcb = () -> @show(loss_sc(x_train, sc_train))
opt = ADAM(0.01)
@epochs 3000 Flux.train!(loss_sc, params(m2), dataset, opt, cb = throttle(evalcb, 10))
```

```julia
record_loss_n_train = Float64[]
record_loss_n_valid = Float64[]

dataset = [(x_train, T_train, ap_train, b_train, y_train)]
evalcb = () -> (push!(record_loss_n_train,loss(x_train, T_train ,ap_train, b_train, y_train).data),
    push!(record_loss_n_valid,loss(x_valid, T_valid ,ap_valid, b_valid, y_valid).data)) #@show(loss_be(x_train, be_train),loss_be(x_valid, be_valid))

opt = ADAM(0.0001)
@epochs 10000 Flux.train!(loss, params(Ae,m1,m2), dataset, opt, cb = throttle(evalcb, 10))
```
