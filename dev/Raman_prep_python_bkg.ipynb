```python
%matplotlib inline

import rampy as rp
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import h5py

import os

import sklearn
import sklearn.model_selection as model_selection
```

# Importing Data  + Baseline correction

```python
my_liste = pd.read_excel("Database_IPGP.xlsx", "raman")
my_liste.head()

# Run
nb_exp = my_liste.shape[0]
roi_glass = my_liste.loc[:,"lb1":"hb2"]
scaling_factor = 1000.   

x = np.arange(400.,1250.,1.0) # our real x axis, for resampling
spectra = np.ones((len(x),nb_exp))
spectra_long = np.ones((len(x),nb_exp))

xmin = np.ones((nb_exp,1))

for i in range(nb_exp):
    file_name, file_extension=os.path.splitext(my_liste.loc[i,"nom"])
    
    if file_extension == ".txt":
        sp = np.genfromtxt("./raman/"+my_liste.loc[i,"nom"],skip_header=1)
    elif file_extension == ".csv":
        sp = np.genfromtxt("./raman/"+my_liste.loc[i,"nom"],skip_header=1,delimiter=",")
    else:
        raise InputError("Unsupported file extension")
        
    sp = rp.flipsp(sp)
    
    print("Spectra %i, checking array size: %i" % (i,sp.shape[0]))
    
    #bir = np.array([[roi_glass.iloc[i,2],roi_glass.iloc[i,3]]])
    idx_min = np.where(sp[:,1] == np.min(sp[(800<= sp[:,0])&(sp[:,0]<=1000),1]))[0][0]
    xmin[i] = sp[idx_min,0]
    
    bir = np.array([[xmin[i]-5,xmin[i]+5],[roi_glass.iloc[i,2],roi_glass.iloc[i,3]]])
    
    #bir = np.array([[roi_glass.iloc[i,2],roi_glass.iloc[i,3]]])
    
    
    y_bas, bas = rp.baseline(sp[:,0],sp[:,1],bir,"poly",polynomial_order=1)

    y_resampled = rp.resample(sp[:,0], y_bas.flatten(), x)
    
    _, y_long, ese_long = rp.tlcorrection(x, y_resampled,23.0,my_liste.loc[i,"laserfreq"],normalisation="intensity")
    
    spectra_long[:,i] = y_long
    
    plt.figure()
    plt.subplot(1,2,1)
    plt.plot(sp[:,0],sp[:,1],"k-")
    plt.plot(sp[:,0],bas,"b-")
    
    plt.subplot(1,2,2)
    plt.plot(x,y_long,"r-")
    
```

```python
X = my_liste.loc[:,["sio2","al2o3","na2o","k2o"]]/100.0
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X,spectra_long.T,test_size=0.3)
```

```python
with h5py.File('./NKAS_Raman.hdf5', 'w') as f:
    f.create_dataset('X_raman_train', data=X_train)
    f.create_dataset('X_raman_test', data=X_test)
    f.create_dataset('y_raman_train', data=y_train)
    f.create_dataset('y_raman_test', data=y_test)
```
